JSON ITEMS AVEC PIECES

```python
df_items = df.copy()
# create a list of our conditions
conditions = [(df_items.index <= 11), #cuisine
              (df_items.index > 11) & (df_items.index <= 24), #salon
              (df_items.index > 24) & (df_items.index <= 27), #sam
              (df_items.index > 27) & (df_items.index <= 36), #chambre
              (df_items.index > 36) & (df_items.index <= 42), #bureau
              (df_items.index > 42) & (df_items.index <= 49), #divers-jardin

            ]


# create a list of the values we want to assign for each condition
values = ['cuisine', 'salon', 'salle à manger', 'chambre','bureau', 'divers-jardin']

# create a new column and use np.select to assign values to it using our lists as arguments
df_items['pièce'] = np.select(conditions, values)

#sort value
#df_items = df_items.sort_values(by=['item'], ascending=True).reset_index(drop=True)




```

```python
pieces = []

for p in pieces_lower :
    pieces.append(p.capitalize())

pieces   
```

SLUG PIECES

```python
slug = []
for i in pieces :
    slug.append(slugify(i))

slug
```

CREATION DU DF PIECES

```python
df_pièces = pd.DataFrame(list(zip(pieces,slug)),
                         columns=['pieces', 'slug'])
```
