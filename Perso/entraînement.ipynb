```python
# Les enfants d'une balise sont disponibles dans une liste appelée .contents:

head_tag = soup.head
head_tag.contents
```

```python
# Si une balise n'a qu'un seul enfant et que cet enfant est un NavigableString, l'enfant est rendu disponible en tant que .string:
title_tag.string
```

```python
# S'il y a plus d'une chose à l'intérieur d'une balise, vous pouvez toujours regarder uniquement les chaînes. Utilisez le .stringsgénérateur :
for string in soup.strings:
    print(repr(string))
```

```python
#Considérez un document simple comme celui-ci :
#<a><b>text1</b><c>text2</c></b></a>
```

```python
sibling_soup = bs("<a><b>text1</b><c>text2</c></b></a>", 'html.parser')
print(sibling_soup.prettify()) # sans le print nous n'avons pas l'indentation et ler etour à la ligne comme ci-dessous
```

Vous pouvez utiliser .next_siblinget .previous_siblingpour naviguer entre les éléments de page qui se trouvent au même niveau de l'arborescence d'analyse :

La balise "b" a un .next_sibling, mais non .previous_sibling, car il n'y a rien avant la balise "b" au même niveau de l'arborescence . Pour la même raison, la balise "c" a un .previous_sibling mais non .next_sibling:

```python
sibling_soup.b.next_sibling
```

```python
sibling_soup.c.previous_sibling
```

```python
# Les chaînes « text1 » et « text2 » ne sont pas sœurs, car elles n'ont pas le même parent :

print(sibling_soup.b.string)
print(sibling_soup.c.string)
```
